A coluna model tem muitos valores únicos, que não deve ter representantes o suficiente para o modelo aprender.
O mesmo acontece com transmission. Nesse caso é possível dividir a informação: tipo de transmissão (5) e quantidade de marchas (ordem crescente).

```python
#extrair os dígitos da coluna transmission
#alguns casos de transmission não tem número. o numpy não aceita NAN como int, apenas como float, por isso vou converter para float.

df['gears'] = df['transmission'].str.extract(r'(\d+)').astype('float')
```

```python
#retirar os valores da coluna transmission
df['transmission'] = df['transmission'].str.replace(r'(\d+)', '', regex=True) #eu preciso dizer que está procurando uma expressão regular.
df.describe(exclude="number")
```

```python
df["fuel_type"].value_counts()
```

```python
fuel = {
    'X':'reg_gasoline',
    'Z':'premium_gasoline',
    'D':'diesel',
    'E':'ethanol',
    'N': 'natural_gas',
}

df['fuel_type'] = df['fuel_type'].map(fuel)
```

```python
df["fuel_type"].value_counts()
```

```python
# carros com gás natural não possuem representatividade suficiente para o aprendizado por poarte do modelo.
df = df.drop(df[df["fuel_type"] == "natural_gas"].index)
```

```python
#checar valores ausentes 
df.isnull().sum()
```

```python
#coluna model tem 3522 valores únicos, pode dificultar a aprendizado do modelo 
# a coluna mpg que replica km, utilizando apenas outra métrica
df=df.drop(columns=['combined_mpg', 'model']) 
```

```python

df_tratados['gears'].unique()
```

```python
#carros com transmissão av possui uma infinita relação de transmissão entre o motor e as rodas.
#para fins de padronização considerarei como 11 
df_tratados.loc[df_tratados['transmission'] == 'AV', 'gears'] = 11
#crosstab para ver a relação das marchas com a transmissão
pd.crosstab(df_tratados['gears'], df_tratados['transmission'], dropna=False)
```

```python
# Definir os bins e labels
bins = [0, 6, 16]  # Limites dos grupos 
labels = ['até 6', '7+']  # Nomes dos grupos

# Criar a nova coluna
df_tratados['gears_grouped'] = pd.cut(
    df_tratados['gears'],
    bins=bins,
    labels=labels,
    right=False #considerar número inicial
)

# Verificar a contagem
df_tratados['gears_grouped'].value_counts()
```

```python
df_tratados['engine_size_l'].unique()	
```

Para agrupar os tamanhos de motor com base na emissão de CO2, vou usar uma abordagem baseada nos valores típicos de emissão associados a diferentes categorias de motores.



```python
# Definir os bins e labels
# Definir os bins e labels
bins = [-np.inf, 1.9 , 3, 5, np.inf]  # Limites dos grupos 
labels = ['up_to_1.9', 'up_to_3', 'up_to_5', 'more_than_5']  # Nomes dos grupos


# Criar a nova coluna
df_tratados['engine_size_l_grouped'] = pd.cut(
    df_tratados['engine_size_l'],
    bins=bins,
    labels=labels,
)

# Verificar a contagem
df_tratados['engine_size_l_grouped'].value_counts()
```

```python
df_tratados['vehicle_class'].unique()
```

```python
class_map = {
    "Two-seater": "car",
    "Minicompact": "car",
    "Subcompact": "car",
    "Compact": "car",
    "Mid-size": "car",
    "Full-size": "car",
    "Station wagon: Small": "station_wagon",
    "Station wagon: Mid-size": "station_wagon",
    "Pickup truck: Small": "pickup_truck",
    "Pickup truck: Standard": "pickup_truck",
    "Sport utility vehicle": "suv",
    "Sport utility vehicle: Small": "suv",
    "Sport utility vehicle: Standard": "suv",
    "Minivan": "van",
    "Van: Cargo": "van",
    "Van: Passenger": "van",
    "Special purpose vehicle": "special_purpose",
}

df_tratados["vehicle_class_grouped"] = df_tratados["vehicle_class"].map(class_map)
```

```python
df_tratados['cylinders'].unique()
```

Sugestão de agrupamento cilindros: baseado em emissões e praticidade):
* 2, 3, 4 cilindros (motores menores, geralmente mais eficientes).
* 6 cilindros são comuns em SUVs e picapes, mas não tão poluentes quanto V8+).
* 8, 10, 12, 16 cilindros (motores maiores, maior consumo e emissões).



```python
# Definir os bins e labels
bins = [-np.inf, 4, 6, 8, np.inf]  # Limites dos grupos 
labels = ['up_to_4', 'up_to_6', 'up_to_8', 'more_than_8']  # Nomes dos grupos

# Criar a nova coluna
df_tratados['cylinders_grouped'] = pd.cut(
    df_tratados['cylinders'],
    bins=bins,
    labels=labels,
)

# Verificar a contagem
df_tratados['cylinders_grouped'].value_counts()
```

```python
# Configuração do LaTeX para CO2
co2_latex = r'CO$_2$'

# Criar figura e eixos
fig, ax = plt.subplots(figsize=(15, 10))

# --- Gráfico de Barras (Emissões de CO2) ---
sns.barplot(
    data=df_tratados,
    x='make',
    y='co2_emissions_g_km',
    ax=ax,
    order=df_tratados.groupby('make')['co2_emissions_g_km'].mean().sort_values(ascending=False).index,
    capsize=0.3,
    zorder=49,
)

# Anotação dos valores (formato 1 decimal)
ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    rotation=90,
    padding=12,
    size=10,
    weight='bold',
    zorder=50,
)

# Configurações do eixo y (esquerdo)
ax.set_yticks([])  # Remove os ticks do eixo y
ax.set_ylabel('')
ax.set_ylim(0, 700)  # Ajuste para os rótulos não saírem do gráfico

# Configurações do eixo x
ax.set_xlabel('')
ax.tick_params(axis='x', rotation=90)

# Título
ax.set_title(f'Emissão de {co2_latex} por fabricante (g/km) x Quantidadede Veículos', color='dimgrey', weight='bold', size=15)

# Cores baseadas no valor de CO2 (gradiente)
norm = mcolors.Normalize(
    vmin=df_tratados['co2_emissions_g_km'].min(),
    vmax=df_tratados['co2_emissions_g_km'].max(),
)
cmap = cm.RdYlGn_r
for barra in ax.patches:
    barra.set_color(cmap(norm(barra.get_height())))

# --- Segundo Eixo (Countplot) ---
ax2 = ax.twinx()  # Cria um eixo y compartilhando o mesmo x

# Gráfico de contagem (countplot)
sns.countplot(
    data=df_tratados,
    x='make',
    ax=ax2,
    order=df_tratados['make'].value_counts().index,  # Ordena pela contagem
    color='skyblue',  # Cor diferente para diferenciar
    alpha=0.5,  # Transparência para não sobrepor totalmente,
    zorder=0
)

# Configurações do eixo y direito (contagem)
ax2.set_ylabel('Quantidade de veículos', color='skyblue')
ax2.tick_params(axis='y', labelcolor='skyblue')

# remover o grid
ax2.grid(False)
sns.despine(ax=ax)

#remover o box(spine)
ax.spines['top'].set_visible(False) 

# Ajuste de layout
plt.tight_layout()
plt.show()
```

O pairplot (do Seaborn) é um gráfico que mostra:

Na diagonal: Distribuições de cada variável (usando KDE - Kernel Density Estimate)

Fora da diagonal: Relações entre pares de variáveis (scatter plots)

Quando usado com hue='Gender', diferencia visualmente os gêneros

O pairplot ajuda a identificar:

* Padrões ou tendências nas relações entre variáveis

* Formas das distribuições das variáveis

* Possíveis outliers ou anomalias nos dados

```python
sns.pairplot(
    df_tratados,
    kind='reg',
    diag_kind='kde',
    plot_kws={'line_kws': {'color':'red'}, 'scatter_kws':{'alpha': 0.25}}
)
plt.show()
```
